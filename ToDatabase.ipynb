```python
import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine
from credentials import username
from credentials import password
from  psycopg2 import connect
```

```python
import_ = 'Resources/usretechnicalpotential(clean).csv'
import_2='Resources/energy_consumption.csv'
df = pd.read_csv(import_)
renewabledf= df.set_index('State')
renewable_sum=renewabledf.sum(axis=1)
renewabledf=pd.DataFrame(renewable_sum)
renewabledf=renewabledf.rename(columns={0:'total_renewable'})
energydf=df.merge(renewabledf,on='State')
energydf=energydf[['State','total_renewable']]
energydf=energydf.rename(columns={'State':'state'})
energydf

```

```python
df2 = pd.read_csv(import_2)
df2=df2.rename(columns={'Total Energy Consumed 2017 (10^6 btu/capita)':'energy_consumed_btu','State_Name':'state', 'State Population 2017':'population', 'Energy_Consumed_GWh': 'energy_consumed_gwh'  })
df2.head()
```

```python
rds_connection_string = f'{username}:{password}@localhost:5433/etl_db'
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
energydf.to_sql(name='state_renewable', con=engine, if_exists='append', index=False)
```

```python
df2.to_sql(name='state_consumption', con=engine, if_exists='append', index=False)
```
