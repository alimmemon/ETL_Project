```python
# import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# read csv file
energy_consumed_src = "TotalEnergyConsumedperCapitaStateRankings.csv"
state_population_src = "nst-est2019-alldata.csv"
state_src = "states.csv"


# create the data frame 
nrg_cons_df = pd.read_csv(energy_consumed_src)
pop_df = pd.read_csv(state_population_src)
states_df = pd.read_csv(state_src)
pop_df = pop_df[['DIVISION','STATE','NAME','POPESTIMATE2017']]
pop_df.head(50)
```

```python
#pop_df['DIVISION'] = pd.to_numeric(pop_df['DIVISION'], errors='ignore')
#pop_df['DIVISION'] = pop_df['DIVISION'].astype(int)
pop_df = pop_df[pop_df['STATE'] != 0]
pop_df.head(50)
```

```python
nrg_cons_df = nrg_cons_df.merge(states_df,left_on='State', right_on='State_Abbr')
nrg_cons_df = nrg_cons_df.drop(columns = ['Rank', 'State_Abbr','Note: Rankings are based on the full source data values.'])
nrg_cons_df.head(50)
```

```python
Total_Energy = pd.merge(pop_df, nrg_cons_df, left_on='NAME', right_on='State_Name')
#Conversion: 1 gigawatt hour (GWh) = 3,412,141,632.81 BTU - British Thermal Unit (btu)
gwh_to_btu = 3412141632.81 
Total_Energy['Energy_Consumed_GWh'] = (Total_Energy['Total Energy Consumed per Capita, million Btu']*1000000*Total_Energy['POPESTIMATE2017'])/gwh_to_btu
Total_Energy = Total_Energy.rename(columns={'POPESTIMATE2017':'State Population 2017','Total Energy Consumed per Capita, million Btu':'Total Energy Consumed 2017 (10^6 btu/capita)'})
Total_Energy = Total_Energy[['State_Name','State Population 2017','Total Energy Consumed 2017 (10^6 btu/capita)','Energy_Consumed_GWh']]
Total_Energy.head(50)
```

```python
consumed_energy_csv = "energy_consumption.csv"
Total_Energy.to_csv(consumed_energy_csv,index=False)
```
